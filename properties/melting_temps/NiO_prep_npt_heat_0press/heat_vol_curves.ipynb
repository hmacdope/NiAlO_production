```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = np.loadtxt("./temp.dat")
vol = np.loadtxt("./vol.dat")
```

```python
plt.plot(temp[:,1])
```

```python
plt.plot(vol[:,1])
```

```python
fig,ax = plt.subplots()
ax.plot(temp[:,1], color='red')
ax2 = ax.twinx()
ax2.plot(vol[:,1], color='blue')
ax.axhline(2072, color='black', label='experimental')
# ax.axhline(1042, color='grey', label='EAM')
fig.legend(["temp", "experimental mt", 'EAM published mt', "volume"])
ax.set_xlabel("step")
ax.set_ylabel("temp (K)")
ax2.set_ylabel("volume $(A^3)$")
plt.savefig("melting_curve.png")
```

```python
fig,ax = plt.subplots()
ax.scatter(temp[:,1], vol[:,1], s=0.01)
ax.set_ylabel("temp (K)")
ax.set_xlabel("volume $(A^3)$")
ax.axvline(2072, color='black', label='experimental')
# ax.axvline(1042, color='grey', label='EAM')
plt.savefig("melt_trace.png")
```
