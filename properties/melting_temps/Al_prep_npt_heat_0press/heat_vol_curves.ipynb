```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = np.loadtxt("./temp.dat")
vol = np.loadtxt("./vol.dat")
```

```python
plt.plot(temp[:,1])
```

```python
plt.plot(vol[:,1])
```

```python
fig,ax = plt.subplots()
ax.plot(temp[:,1], color='red')
ax2 = ax.twinx()
ax2.plot(vol[:,1], color='blue')
ax.axhline(933, color='grey', label='experimental')
ax.axhline(1042, color='black', label='EAM')
fig.legend(["temp", "experimental", 'EAM', "volume"])
```

```python
fig,ax = plt.subplots()
ax.scatter(temp[:,1], vol[:,1])
```
