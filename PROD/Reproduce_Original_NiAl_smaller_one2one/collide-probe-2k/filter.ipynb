```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
```

```python
!pwd
```

```python
ke = np.loadtxt("ke.dat")
```

```python
axis = ke[:,0]
ke_vals = ke[:,1]
```

```python
filtered = scipy.signal.savgol_filter(ke_vals, 100, 3 )
```

```python
ke_vals
```

```python
filtered
```

```python
plt.plot(axis, ke_vals)
plt.plot(axis, filtered)
```

```python
peaks = scipy.signal.find_peaks(-filtered, width=20, distance=2000)
```

```python
plt.plot(axis, -filtered)
```

```python
peaks
```

```python
plt.plot(axis, ke_vals)
plt.plot(axis, filtered)
for i in peaks[0]:
    plt.axvline(i)
plt.savefig("min_ke_plot.png")
```
